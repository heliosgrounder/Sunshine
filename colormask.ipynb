```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv.imread('90100002.jpg')
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
hsv_img1 = cv.cvtColor(img1, cv.COLOR_RGB2HSV)
img2 = cv.imread('90100001.jpg')
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
hsv_img2 = cv.cvtColor(img2, cv.COLOR_RGB2HSV)
img3 = cv.imread('90100022.jpg')
img3 = cv.cvtColor(img3, cv.COLOR_BGR2RGB)
hsv_img3 = cv.cvtColor(img3, cv.COLOR_RGB2HSV)
```

```python
plt.subplot(1, 3, 1)
plt.imshow(hsv_img1)
plt.subplot(1, 3, 2)
plt.imshow(hsv_img2)
plt.subplot(1, 3, 3)
plt.imshow(hsv_img3)
plt.show()
```

```python
first_color = (0, 0, 0)
second_color = (200, 100, 100)
mask = cv.inRange(hsv_img1, first_color, second_color)
result = cv.bitwise_and(img1, img1, mask = mask)
plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.subplot(1, 2, 2)
plt.imshow(mask)
plt.show()
```
